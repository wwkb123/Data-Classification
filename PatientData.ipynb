```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('output.txt', sep="\t", header=None)
```

```python
df.head()
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
```

```python
df.drop('PatientPopulationPercentageBelowPoverty',axis=1,inplace=True)
```

```python
df.head()
```

```python
import random
ssn_list = []
for x in range(0,900):
    random.seed(x)
    first = random.randint(100,999)
    mid = random.randint(10,99)
    last = random.randint(1000,9999)
    ssn = str(first)+"-"+str(mid)+"-"+str(last)
    ssn_list.append(ssn)
```

```python
df['PatientSSN'] = ssn_list
```
